# Example code and plots shown in the poster

```python
import pygmt as pg
```

```python
topo = pg.datasets.load_earth_relief(resolution="30m")
topo
```

```python
fig = pg.Figure()
fig.basemap(region=[-180, 180, -70, 70], projection="M30i", B="f")
fig.grdimage(topo, cmap="abyss")
fig.show(width=1000)
```
