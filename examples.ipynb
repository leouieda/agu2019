# Example code and plots shown in the poster

```python
import pygmt
```

## A first figure

```python
# The Figure controls all plotting
fig = pygmt.Figure()
# Draw the frame and set region and projection
fig.basemap(region=[-100, -50, 5, 25], projection="M20c", frame=True)
# Plot shorelines
fig.coast(shorelines=True)
# savefig is the same as matplotlib
fig.savefig("central-america.png")
# Display the figure
fig.show()
```

## Plotting coastlines

```python
oahu = [-158.3, -157.6, 21.2, 21.8]
fig = pygmt.Figure()
for res in ["c", "l", "i", "h", "f"]:
    fig.coast(resolution=res, shorelines="2p", 
              region=oahu, projection="M10c")
    fig.shift_origin(xshift="10c")
fig.savefig("coastlines.png")
fig.show()
```
